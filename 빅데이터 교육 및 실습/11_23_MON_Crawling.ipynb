```python
from konlpy.tag import Hannanum
from collections import Counter

tagger = Hannanum()

text = "대한민국은 민주공화국이다. 대한민국의 주권은 국민에게 있고, 국민은 그 대표자나 국민투표에 의하여 주권을 행사한다."
result = tagger.morphs(text)

c = Counter(result)
print(c.most_common(5))
```

```python
from konlpy.tag import Okt
import re
with open("/home/piai/Downloads/news.txt","r", encoding = "utf-8") as f:
    content = f.read()
    
p = re.compile("[\Wa-zA-Z0-9_]+")
content = re.sub(p," ",content)

tagger=Okt()
morphs = tagger.pos(content)
print(morphs)
```

```python
words= []
for word, pos, in morphs:
    if pos == 'Noun':
        words.append(word)
        
print(words)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
data= pd.read_csv("/home/piai/movies_metadata.csv", low_memory = False)
data = data.iloc[:20000]
data.head()
```

```python
data['overview'] = data['overview'].fillna("")
print(data.shape)
```

```python
tfidf = TfidfVectorizer(stop_words = 'english')
tfidf_matrix = tfidf.fit_transform(data['overview'])
print(tfidf_matrix.shape)
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
print(cosine_sim)
```

```python
cosine_sim = linear_kernel(tfidf_matrix,tfidf_matrix)

title = "The Dark Knight Rises"
title2index = pd.Series(data=data.index, index=data['title']).drop_duplicates()
index = title2index[title]

sim_scores = list(enumerate(cosine_sim[index]))

sim_scores.sort(key= lambda x:x[1], reverse = True)
top10 = sim_scores[1:11]
top10_index = [x[0] for x in top10]
print(top10_index)
print(data['title'].iloc[top10_index])
```
