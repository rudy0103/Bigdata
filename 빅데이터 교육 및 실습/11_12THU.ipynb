```python
left=pd.DataFrame({'key1':['K0','K1','K2','K3'],
                   'key2':['K0','K1','K0','K1'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})
right=pd.DataFrame({'key1':['K0','K1','K2','K3'],
                   'key2':['K0','K0','K0','K0'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']})
```

```python
result = pd.merge(left,right,on=['key1','key2'])
```

```python
result
```

```python
result = pd.merge(left,right,on=['key1','key2'],how = 'left')
result
```

```python
result = pd.merge(left,right, on =['key1','key2'],how = 'right')
result
```

```python
result = pd.merge(left,right, on =['key1','key2'],how = 'outer')
result
```

```python
result = pd.merge(left,right, on =['key1','key2'],how = 'outer',indicator=True)
result
```

```python
import pandas as pd
import numpy as np

pop_seoul = pd.read_csv("/home/piai/bigdata/data/Handling/서울인구통계.txt",sep='\t',engine='python',thousands=',', encoding = "utf-8")

pop1 = pop_seoul[['자치구','인구','인구.3','인구.6','65세이상고령자']]
pop_seoul=pop1.drop([0,1])
pop_seoul.head()
```

```python
pop_seoul.rename(columns={pop_seoul.columns[0]:'구별',
                          pop_seoul.columns[1]:'인구수',
                          pop_seoul.columns[2]:'한국인',
                          pop_seoul.columns[3]:'외국인',
                          pop_seoul.columns[4]:'고령자'
                         },inplace=True)
pop_seoul.head()
```
